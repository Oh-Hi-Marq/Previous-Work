### Problem 4. 
Use symbolic python package to define a function $f(x) = 2x-2e^x-3x^3$ and find it's derivatives $f'(x)$, and $f''(x)$. Finally plot all three functions on a same graph.

```python
# Making x a usable variable.
x=sym.Symbol('x')

# Defining our function.
def f(x): return 2*x-2*exp(x)-3*x**3
f(x)
```

```python
# Defining the first derivative of our function.
def df(x): return sym.diff(f(x))
df(x)
```

```python
# Defining the second derivative of our function.
def ddf(x): return sym.diff(df(x))
ddf(x)
```

```python
# Plotting the functions.
sym.plot(f(x),df(x),ddf(x))
# I tried to change the colors for each function, but I failed.
```

### Problem 8
A distribution (here, it is one of the numerical columns) in which the values of mean, median, and mode coincide (i.e., Mean = Median = Mode) is known as a normal distribution (a perfect data set). It has a bell-shaped curve when plotted and is also symmetric about the mean. Conversely, when mean, median, and mode values are not equal, the distribution is asymmetrical or skewed. In a moderately skewed or asymmetrical distribution, a significant relationship exists among these three measures of central tendencies. In such distributions, the distance between the mean and median is about one-third of the mean and Mode distance. Karl Pearson, a renowned statistician, expressed this relationship as:
$Mode - 3 \times \,\,median + 2\times \,\, mean =0$. 
Find Mode - 3 Median + 2 Mean for all the numerical columns. Symmetric data is hard to come by. If this number is close to 0 for a column, the column data is close to being symmetric, and if it is far from 0 ( positive or negative), the column data is asymmetric.
