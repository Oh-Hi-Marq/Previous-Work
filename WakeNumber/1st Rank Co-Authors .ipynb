```python
from scholarly import scholarly
import ast
import pandas as pd
```

```python
gs_wake = scholarly.fill(scholarly.search_author_id('jNUs6LsAAAAJ'), sections=['coauthors'])

gs_wake
```

```python
def filter1(pair):
    wanted_keys = ['coauthors']
    key, value = pair
    if key in wanted_keys:
        return True  # keep pair in the filtered dictionary
    else:
        return False  # filter pair out of the dictionary
```

```python
filter1_wake = dict(filter(filter1, gs_wake.items()))

filter1_wake
```

```python
rep = str(filter1_wake).replace("{'coauthors': ",'').replace("]}",'').replace("'source': <AuthorSource.CO_AUTHORS_LIST: 'CO_AUTHORS_LIST'>,",'') + ']'

print(rep)
```

```python
list_wake = ast.literal_eval(rep)
```

```python
def test(lsts, key):
  return [x.get(key) for x in lsts]
```

```python
rank_1 = {test(list_wake, 'name')[i]: test(list_wake, 'scholar_id')[i] for i in range(len(test(list_wake, 'name')))}

rank_1
```

```python
wake_1 = pd.DataFrame.from_dict(rank_1, orient='index').reset_index()
wake_1.columns = ['Name', 'ID']
wake_1
```

```python
wake_1['Wake #'] = 1
wake1 = wake_1.set_index('Wake #')
```
